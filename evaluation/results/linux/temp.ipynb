```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for the privacy metrics
data_privacy = {
    "Model": ["BeGreat", "CTGAN", "LLaMa-3"],
    "NND_Ratio_1st": [0.7223, 0.7283, 0.7234],
    "NND_Ratio_2nd": [0.8898, 0.9460, 0.0437],
    "Median_Dist_1st": [1.1470, 1.8081, float('inf')],
    "Median_Dist_2nd": [7.6622, 1.1345, 1.581],
    "Hitting_Rate_1st": [0.0144, 0.0144, 0.0172],
    "Hitting_Rate_2nd": [0.0, 0.0, 0.0],
    "Epsilon_Risk_1st": [0.3849, 0.3381, 0.0805],
    "Epsilon_Risk_2nd": [0.0002, 0.2122, 0.1729],
    "Attr_Disclosure_1st": [0.8044, 0.7993, 0.8540],
    "Attr_Disclosure_2nd": [0.1498, 0.2041, 0.3977]
}

# Create a DataFrame
df_privacy = pd.DataFrame(data_privacy)

# Normalizing the values
df_privacy_normalized = df_privacy.copy()
columns_to_normalize = df_privacy.columns[1:]

for column in columns_to_normalize:
    max_value = df_privacy[column].replace([np.inf, -np.inf], np.nan).max()
    df_privacy_normalized[column] = df_privacy[column] / max_value

# Setting the positions and width for the bars
pos = list(range(len(df_privacy['Model'])))
width = 0.1

# Plotting the first dataset with normalized values
fig1, ax1 = plt.subplots(figsize=(10, 6))
plt.rcParams.update({'font.size': 14})

# Create bars for each metric with new colors
ax1.bar(pos, df_privacy_normalized['NND_Ratio_1st'], width, color='#1f77b4', label='Nearest Neighbour Distance Ratio')
ax1.bar([p + width for p in pos], df_privacy_normalized['Median_Dist_1st'], width, color='#ff7f0e', label='Median Distance to Closest Record')
ax1.bar([p + width*2 for p in pos], df_privacy_normalized['Hitting_Rate_1st'], width, color='#2ca02c', label='Hitting Rate')
ax1.bar([p + width*3 for p in pos], df_privacy_normalized['Epsilon_Risk_1st'], width, color='#d62728', label='Epsilon Identifiability Risk')
ax1.bar([p + width*4 for p in pos], df_privacy_normalized['Attr_Disclosure_1st'], width, color='#9467bd', label='Attribute Disclosure Risk')

# Set the y axis label
ax1.set_ylabel('Normalized Values')

# Set the chart's title
ax1.set_title('Comparison of Normalized Privacy Metrics for Synthetic Data Models - First Dataset')

# Set the position of the x ticks
ax1.set_xticks([p + 2 * width for p in pos])

# Set the labels for the x ticks
ax1.set_xticklabels(df_privacy_normalized['Model'])

# Setting the x-axis and y-axis limits
ax1.set_xlim(min(pos)-width, max(pos)+width*5)
ax1.set_ylim([0, 1.1])

# Adding the legend and showing the plot
ax1.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax1.grid()
plt.show()

# Plotting the second dataset with normalized values
fig2, ax2 = plt.subplots(figsize=(10, 6))
plt.rcParams.update({'font.size': 14})

# Create bars for each metric with new colors
ax2.bar(pos, df_privacy_normalized['NND_Ratio_2nd'], width, color='#1f77b4', label='Nearest Neighbour Distance Ratio')
ax2.bar([p + width for p in pos], df_privacy_normalized['Median_Dist_2nd'], width, color='#ff7f0e', label='Median Distance to Closest Record')
ax2.bar([p + width*2 for p in pos], df_privacy_normalized['Hitting_Rate_2nd'], width, color='#2ca02c', label='Hitting Rate')
ax2.bar([p + width*3 for p in pos], df_privacy_normalized['Epsilon_Risk_2nd'], width, color='#d62728', label='Epsilon Identifiability Risk')
ax2.bar([p + width*4 for p in pos], df_privacy_normalized['Attr_Disclosure_2nd'], width, color='#9467bd', label='Attribute Disclosure Risk')

# Set the y axis label
ax2.set_ylabel('Normalized Values')

# Set the chart's title
ax2.set_title('Comparison of Normalized Privacy Metrics for Synthetic Data Models - Second Dataset')

# Set the position of the x ticks
ax2.set_xticks([p + 2 * width for p in pos])

# Set the labels for the x ticks
ax2.set_xticklabels(df_privacy_normalized['Model'])

# Setting the x-axis and y-axis limits
ax2.set_xlim(min(pos)-width, max(pos)+width*5)
ax2.set_ylim([0, 1.1])

# Adding the legend and showing the plot
ax2.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax2.grid()
plt.show()
```
