```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('datasets/data.csv')

print(df.shape[0])

plt.figure(figsize=(24, 10))
sns.countplot(x='Disease', data=df)
plt.title('Distribution of Diseases')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
symptoms = df[['Disease', 'Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']]
symptoms = symptoms.replace({'Yes': 1, 'No': 0})

symptoms_melted = symptoms.melt(id_vars='Disease', var_name='Symptom', value_name='Presence')
heatmap_data = symptoms_melted.groupby(['Disease', 'Symptom']).Presence.sum().unstack()

plt.figure(figsize=(20, 20))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
plt.title('Heatmap of Symptoms Across Diseases')
plt.show()
```

```python
plt.figure(figsize=(24, 10))
sns.countplot(x='Disease', hue='Outcome Variable', data=df)
plt.title('Outcome Variable Distribution by Disease')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

```

```python
plt.figure(figsize=(28, 10))
sns.countplot(x='Disease', hue='Blood Pressure', data=df)
plt.title('Blood Pressure Status by Disease')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(28, 10))
sns.countplot(x='Disease', hue='Cholesterol Level', data=df)
plt.title('Cholesterol Level Status by Disease')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(28, 10))
sns.pairplot(df[['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']].replace({'Yes': 1, 'No': 0}))
plt.title('Pairwise Relationships Between Symptoms')
plt.show()
```
